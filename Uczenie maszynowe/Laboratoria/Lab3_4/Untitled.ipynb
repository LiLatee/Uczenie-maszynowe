```python
import numpy as np
# data to plot
n_groups = 2
accuracy_trees = [1-0.05, 1-0.163]
accuracy_rules = [1-0.057, 1-0.139]


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, accuracy_trees, bar_width,
alpha=opacity,
color='black',
label='Drzewo')

rects2 = plt.bar(index + bar_width, accuracy_rules, bar_width,
alpha=opacity,
color='#9c9a9a',
label='Reguły')

ax.set_ylim(ymin=0.8)
plt.ylabel('Trafność klasyfikacji')
plt.xlabel('Zbiory')
plt.xticks(index + bar_width/2, ('VOTE', 'CRX'))
plt.legend()


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2.,
                1.01*height,
                '%.3f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.savefig('./Wykresy/Lab4/Zad2/zad2_trafnosc.png')
plt.show()
```
